```python
import mediapipe as mp
import cv2
import os

mp_holistic = mp.solutions.holistic.Holistic(
    min_detection_confidence=0.1,
    min_tracking_confidence=0.1,
    static_image_mode=True
)
mp_drawing = mp.solutions.drawing_utils

for frame_file in [f"/content/frame_{i}.jpg" for i in [0, 10, 20, 30, 40]]:
    if os.path.exists(frame_file):
        frame = cv2.imread(frame_file)
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = mp_holistic.process(frame_rgb)
        frame_bgr = cv2.cvtColor(frame_rgb, cv2.COLOR_RGB2BGR)
        hand_detected = False
        if results.right_hand_landmarks:
            mp_drawing.draw_landmarks(frame_bgr, results.right_hand_landmarks, mp.solutions.holistic.HAND_CONNECTIONS)
            hand_detected = True
            print(f"Right hand detected in {os.path.basename(frame_file)}")
        if results.left_hand_landmarks:
            mp_drawing.draw_landmarks(frame_bgr, results.left_hand_landmarks, mp.solutions.holistic.HAND_CONNECTIONS)
            hand_detected = True
            print(f"Left hand detected in {os.path.basename(frame_file)}")
        if not hand_detected:
            print(f"No hand detected in {os.path.basename(frame_file)}")
        cv2.imwrite(f"/content/annotated_{os.path.basename(frame_file)}", frame_bgr)
        print(f"Saved annotated_{os.path.basename(frame_file)}")
mp_holistic.close()
```

```python
import mediapipe as mp
import cv2
import numpy as np
import os
import pickle
from google.colab import drive

mp_holistic = mp.solutions.holistic.Holistic(
    min_detection_confidence=0.1,
    min_tracking_confidence=0.1,
    static_image_mode=True
)

def extract_landmarks_from_frames(video_name, frame_dir, output_dir):
    output_file = os.path.join(output_dir, video_name, 'landmarks_complete.pkl')
    print(output_file)
    if os.path.exists(output_file):
        print(f"Skipping {video_name}")
        return

    landmarks = []
    frame_files = sorted([f for f in os.listdir(frame_dir) if f.startswith("frame_") and f.endswith(".jpg")])
    for frame_file in frame_files:
        frame_path = os.path.join(frame_dir, frame_file)
        if os.path.exists(frame_path):
            frame = cv2.imread(frame_path)
            results = mp_holistic.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            frame_landmarks = np.zeros((543, 3))
            if results.pose_landmarks:
                pose = np.array([[lm.x, lm.y, lm.z] for lm in results.pose_landmarks.landmark])
                frame_landmarks[:33, :] = pose
            if results.right_hand_landmarks:
                right_hand = np.array([[lm.x, lm.y, lm.z] for lm in results.right_hand_landmarks.landmark])
                frame_landmarks[468:489, :] = right_hand
                print(f"Right hand detected in {frame_file}")
            if results.left_hand_landmarks:
                left_hand = np.array([[lm.x, lm.y, lm.z] for lm in results.left_hand_landmarks.landmark])
                frame_landmarks[489:510, :] = left_hand
                print(f"Left hand detected in {frame_file}")
            if results.face_landmarks:
                face = np.array([[lm.x, lm.y, lm.z] for lm in results.face_landmarks.landmark])
                frame_landmarks[33:468, :] = face
            landmarks.append(frame_landmarks)
    landmarks = np.array(landmarks)
    os.makedirs(os.path.dirname(output_file), exist_ok=True)
    with open(output_file, 'wb') as f:
        pickle.dump(landmarks, f)
    print(f"Saved {video_name}: Shape {landmarks.shape}, Hands: {np.any(landmarks[:, 468:510, :] != 0)}")

video_name = "pk-hfad-1_1"
frame_dir = "/content"
output_dir = '/content/drive/MyDrive/landmarks_data_corrected'
extract_landmarks_from_frames(video_name, frame_dir, output_dir)
mp_holistic.close()
```

```python
import sign_language_translator as slt
import mediapipe as mp
import cv2
import numpy as np
import os
import pickle
from google.colab import drive

mp_holistic = mp.solutions.holistic.Holistic(
    min_detection_confidence=0.1,
    min_tracking_confidence=0.1,
    static_image_mode=True
)

def extract_landmarks_from_video(video_path, output_dir):
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    output_file = os.path.join(output_dir, video_name, 'landmarks_complete.pkl')
    if os.path.exists(output_file):
        os.remove(output_file)
        print(f"Overwriting existing {output_file}")

    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frames = []
    for i in range(frame_count):
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)
        cv2.imwrite(f"/content/frame_{i}.jpg", frame)
    cap.release()
    print(f"Extracted {len(frames)} frames")

    landmarks = []
    for frame_idx in range(len(frames)):
        frame = frames[frame_idx]
        frame = cv2.resize(frame, (640, 480))
        results = mp_holistic.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        frame_landmarks = np.zeros((543, 3))
        if results.pose_landmarks:
            pose = np.array([[lm.x, lm.y, lm.z] for lm in results.pose_landmarks.landmark])
            frame_landmarks[0:33, :] = pose
        if results.face_landmarks:
            face = np.array([[lm.x, lm.y, lm.z] for lm in results.face_landmarks.landmark])
            frame_landmarks[33:501, :] = face
        if results.right_hand_landmarks:
            right_hand = np.array([[lm.x, lm.y, lm.z] for lm in results.right_hand_landmarks.landmark])
            frame_landmarks[501:522, :] = right_hand
            print(f"Right hand detected in frame {frame_idx}")
        if results.left_hand_landmarks:
            left_hand = np.array([[lm.x, lm.y, lm.z] for lm in results.left_hand_landmarks.landmark])
            frame_landmarks[522:543, :] = left_hand
            print(f"Left hand detected in frame {frame_idx}")
        landmarks.append(frame_landmarks)
    landmarks = np.array(landmarks)
    os.makedirs(os.path.dirname(output_file), exist_ok=True)
    with open(output_file, 'wb') as f:
        pickle.dump(landmarks, f)
    print(f"Saved {video_name}: Shape {landmarks.shape}, Hands: {np.any(landmarks[:, 501:543, :] != 0)}")

video_path = '/content/drive/MyDrive/SL-PSL-RTP/sign_language_translator/assets/video1/videos/pk-hfad-1_1.mp4'
output_dir = '/content/drive/MyDrive/landmarks_data_corrected'
extract_landmarks_from_video(video_path, output_dir)
mp_holistic.close()
```

```python
import sign_language_translator as slt
import mediapipe as mp
import cv2
import numpy as np
import os
import pickle
from google.colab import drive

mp_holistic = mp.solutions.holistic.Holistic(
    min_detection_confidence=0.1,
    min_tracking_confidence=0.1,
    static_image_mode=True
)

def extract_landmarks_from_video(video_path, output_dir):
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    output_folder = os.path.join(output_dir, video_name)
    os.makedirs(output_folder, exist_ok=True)
    output_file = os.path.join(output_folder, 'landmarks_complete.pkl')
    if os.path.exists(output_file):
        os.remove(output_file)
        print(f"Overwriting existing {output_file}")

    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frames = []
    for i in range(frame_count):
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)
        # Optional: Save individual frames for debugging
        # cv2.imwrite(f"/content/{video_name}_frame_{i}.jpg", frame)
    cap.release()
    print(f"Extracted {len(frames)} frames from {video_name}")

    landmarks = []
    for frame_idx in range(len(frames)):
        frame = frames[frame_idx]
        frame = cv2.resize(frame, (640, 480))
        results = mp_holistic.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        frame_landmarks = np.zeros((543, 3))
        if results.pose_landmarks:
            pose = np.array([[lm.x, lm.y, lm.z] for lm in results.pose_landmarks.landmark])
            frame_landmarks[0:33, :] = pose
        if results.face_landmarks:
            face = np.array([[lm.x, lm.y, lm.z] for lm in results.face_landmarks.landmark])
            frame_landmarks[33:501, :] = face
        if results.right_hand_landmarks:
            right_hand = np.array([[lm.x, lm.y, lm.z] for lm in results.right_hand_landmarks.landmark])
            frame_landmarks[501:522, :] = right_hand
            # print(f"Right hand detected in frame {frame_idx} of {video_name}")
        if results.left_hand_landmarks:
            left_hand = np.array([[lm.x, lm.y, lm.z] for lm in results.left_hand_landmarks.landmark])
            frame_landmarks[522:543, :] = left_hand
            # print(f"Left hand detected in frame {frame_idx} of {video_name}")
        landmarks.append(frame_landmarks)
    landmarks = np.array(landmarks)
    with open(output_file, 'wb') as f:
        pickle.dump(landmarks, f)
    print(f"Saved {video_name}: Shape {landmarks.shape}, Hands: {np.any(landmarks[:, 501:543, :] != 0)}")

def process_folder(input_folder, output_folder):
    for item in os.listdir(input_folder):
        item_path = os.path.join(input_folder, item)
        if os.path.isfile(item_path) and item.lower().endswith(('.mp4', '.avi', '.mov')):
            print(f"Processing video: {item_path}")
            extract_landmarks_from_video(item_path, output_folder)
        elif os.path.isdir(item_path):
            print(f"Found subfolder: {item_path}. Processing videos inside.")
            process_folder(item_path, output_folder) # Recursive call for subfolders

# Define the input and output directories
input_video_folder = '/content/drive/MyDrive/SL-PSL-RTP/sign_language_translator/assets/video1/videos'
output_landmarks_folder = '/content/drive/MyDrive/landmarks_data_corrected'

# Create the output folder if it doesn't exist
os.makedirs(output_landmarks_folder, exist_ok=True)

# Process all videos in the input folder and its subfolders
process_folder(input_video_folder, output_landmarks_folder)

# Close the mediapipe holistic model
mp_holistic.close()

print("Finished processing all videos.")
```

```python
def normalize_landmarks(landmarks):
    mean = landmarks.mean(axis=(0, 1), keepdims=True)
    std = landmarks.std(axis=(0, 1), keepdims=True) + 1e-8
    return (landmarks - mean) / std

def pad_landmarks(landmarks, target_frames=100):
    current_frames = landmarks.shape[0]
    if current_frames < target_frames:
        pad = np.zeros((target_frames - current_frames, landmarks.shape[1], landmarks.shape[2]))
        landmarks = np.vstack([landmarks, pad])
    return landmarks[:target_frames]
```

```python
class SignToTextModel(nn.Module):
    def __init__(self, landmark_dim=375, hidden_dim=256, vocab_size=5000, num_layers=2, max_seq_len=50):
        super(SignToTextModel, self).__init__()

        # CNN for feature extraction
        self.cnn = nn.Conv1d(landmark_dim, hidden_dim, kernel_size=3, padding=1)

        # Transformer Encoder
        encoder_layer = nn.TransformerEncoderLayer(d_model=hidden_dim, nhead=8, batch_first=True)  # Set batch_first=True
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        # Transformer Decoder
        decoder_layer = nn.TransformerDecoderLayer(d_model=hidden_dim, nhead=8, batch_first=True)  # Set batch_first=True
        self.decoder = nn.TransformerDecoder(decoder_layer, num_layers=num_layers)

        # Text embedding
        self.text_embedding = nn.Embedding(vocab_size, hidden_dim)

        # Output layer
        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, landmarks, target_text, src_key_padding_mask=None, tgt_key_padding_mask=None):
        # 1. CNN Feature Extraction
        landmarks = landmarks.permute(0, 2, 1)
        features = self.cnn(landmarks)
        features = features.permute(0, 2, 1)
        print(f"Shape of features: {features.shape}") # Add a print statement

        # 2. Transformer Encoding
        encoder_output = self.encoder(features, src_key_padding_mask=src_key_padding_mask) # Pass src_key_padding_mask to encoder

        # 3. Transformer Decoding
        target_embeddings = self.text_embedding(target_text)
        decoder_output = self.decoder(target_embeddings, encoder_output, tgt_key_padding_mask=tgt_key_padding_mask) # Pass tgt_key_padding_mask to decoder

        # 4. Output Layer
        output = self.fc(decoder_output)
        return output
```

```python
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F

class SignTextDataset(Dataset):
    def __init__(self, data, max_frames=300, max_text_len=50):
        self.landmarks = [item[0] for item in data]
        self.texts = [item[1] for item in data]
        self.max_frames = max_frames
        self.max_text_len = max_text_len

        # Tokenization (You'll need to implement or use an existing tokenizer)
        # self.tokenizer = YourTokenizer()  # Replace with your tokenizer
        self.text_indices = [word_tokenize(text) for text in self.texts]

    def __len__(self):
        return len(self.landmarks)

    def __getitem__(self, idx):
        landmarks = torch.tensor(self.landmarks[idx], dtype=torch.float32)
        text = torch.tensor(self.text_indices[idx], dtype=torch.long)

        # Padding/Truncation
        landmarks = self._pad_or_truncate(landmarks, self.max_frames)
        text = self._pad_or_truncate(text, self.max_text_len, value=0)  # 0 for padding

        return landmarks, text

    def _pad_or_truncate(self, sequence, max_len, value=0):
        seq_len = sequence.size(0)
        if seq_len > max_len:
            sequence = sequence[:max_len]
        else:
            padding_size = (0, max_len - seq_len)
            sequence = F.pad(sequence, padding_size, value=value)
        return sequence
```
